```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np

df_data = pd.read_csv("DiplomadosFinal.csv")
```

```python
#Todas os Estabelecimentos
df_data["Estabelecimento"].drop_duplicates()
```

```python
#No total de diplomados – Mestrado | Doutoramento | Licenciatura

licenciatura = df_data.loc[df_data["Nivel_forma"]=="Licenciatura"]["Total"].sum()
mestrado = df_data.loc[df_data["Nivel_forma"]=="Mestrado"]["Total"].sum()
doutoramento = df_data.loc[df_data["Nivel_forma"]=="Doutoramento"]["Total"].sum()

print("Total diplomados Licenciatura: " + str(licenciatura))
print("Total diplomados Mestrado: " +  str(mestrado))
print("Total diplomados Doutoramento: " +  str(doutoramento))
print("Total diplomados: " +  str(licenciatura + mestrado + doutoramento))
```

```python
#No total de diplomados do sexo F p/tipo de curso
licenciatura = df_data.loc[df_data["Nivel_forma"]=="Licenciatura"]["Sexo_m"].sum()
mestrado = df_data.loc[df_data["Nivel_forma"]=="Mestrado"]["Sexo_m"].sum()
doutoramento = df_data.loc[df_data["Nivel_forma"]=="Doutoramento"]["Sexo_m"].sum()

print("Total diplomados sexo F Licenciatura: " + str(licenciatura))
print("Total diplomados sexo F Mestrado: " +  str(mestrado))
print("Total diplomados sexo F Doutoramento: " +  str(doutoramento))
print("Total diplomados sexo F: " +  str(licenciatura + mestrado + doutoramento))
```

```python
#No total de diplomados do sexo M p/tipo de curso
licenciatura = df_data.loc[df_data["Nivel_forma"]=="Licenciatura"]["Sexo_h"].sum()
mestrado = df_data.loc[df_data["Nivel_forma"]=="Mestrado"]["Sexo_h"].sum()
doutoramento = df_data.loc[df_data["Nivel_forma"]=="Doutoramento"]["Sexo_h"].sum()

print("Total diplomados sexo M Licenciatura: " + str(licenciatura))
print("Total diplomados sexo M Mestrado: " +  str(mestrado))
print("Total diplomados sexo M Doutoramento: " +  str(doutoramento))
print("Total diplomados sexo M: " +  str(licenciatura + mestrado + doutoramento))
```

```python
#Nº de diplomados por nacionalidade

estrangeiro = df_data["Estrangeiro"].sum()
portugues = df_data["Portugues"].sum()

print("Total diplomados estrangeiros: " + str(estrangeiro))
print("Total diplomados portugueses: " +  str(portugues))
print("Total diplomados: " +  str(estrangeiro + portugues))
```

```python
#Nº total de diplomados p/concelho

concelhos = df_data["Conselho"].drop_duplicates()

for concelho in concelhos.index:
    newConcelho = df_data.loc[concelho]["Conselho"]
    totalPortugueses = df_data.loc[df_data["Conselho"]==newConcelho]["Portugues"].sum()
    totalEstrangeiros = df_data.loc[df_data["Conselho"]==newConcelho]["Estrangeiro"].sum()
    
    print(newConcelho)
    print("  -Portugueses: " + str(totalPortugueses))
    print("  -Estrangeiros: " + str(totalEstrangeiros))
    
```

```python
#Nº total de diplomados p/distrito

distritos = df_data["Distrito"].drop_duplicates()

for distrito in distritos.index:
    newDistrito = df_data.loc[distrito]["Distrito"]
    totalPortugueses = df_data.loc[df_data["Distrito"]==newDistrito]["Portugues"].sum()
    totalEstrangeiros = df_data.loc[df_data["Distrito"]==newDistrito]["Estrangeiro"].sum()
    
    print(newDistrito)
    print("  -Portugueses: " + str(totalPortugueses))
    print("  -Estrangeiros: " + str(totalEstrangeiros))
```

```python
#Nº total diplomados tipo de curso

licenciatura = df_data.loc[df_data["Nivel_forma"]=="Licenciatura"]["Estrangeiro"].sum()
mestrado = df_data.loc[df_data["Nivel_forma"]=="Mestrado"]["Estrangeiro"].sum()
doutoramento = df_data.loc[df_data["Nivel_forma"]=="Doutoramento"]["Estrangeiro"].sum()

print("Total diplomados Estrangeiros Licenciatura: " + str(licenciatura))
print("Total diplomados Estrangeiros Mestrado: " +  str(mestrado))
print("Total diplomados Estrangeiros Doutoramento: " +  str(doutoramento))
print("Total diplomados Estrangeiros: " +  str(licenciatura + mestrado + doutoramento))

licenciatura = df_data.loc[df_data["Nivel_forma"]=="Licenciatura"]["Portugues"].sum()
mestrado = df_data.loc[df_data["Nivel_forma"]=="Mestrado"]["Portugues"].sum()
doutoramento = df_data.loc[df_data["Nivel_forma"]=="Doutoramento"]["Portugues"].sum()

print("-----------------------------")

print("Total diplomados Portugues Licenciatura: " + str(licenciatura))
print("Total diplomados Portugues Mestrado: " +  str(mestrado))
print("Total diplomados Portugues Doutoramento: " +  str(doutoramento))
print("Total diplomados Portugues: " +  str(licenciatura + mestrado + doutoramento))
```

```python
#Nº total de diplomados área de estudo

areasEnsino = df_data["Cnaef_grande_grupo"].drop_duplicates()

for area in areasEnsino.index:
    areaEstudo = df_data.loc[area]["Cnaef_grande_grupo"]
    totalPortugueses = df_data.loc[df_data["Cnaef_grande_grupo"]==areaEstudo]["Portugues"].sum()
    totalEstrangeiros = df_data.loc[df_data["Cnaef_grande_grupo"]==areaEstudo]["Estrangeiro"].sum()
    
    print(areaEstudo)
    print("  -Portugueses: " + str(totalPortugueses))
    print("  -Estrangeiros: " + str(totalEstrangeiros))
```

```python
#Nº total de diplomados por UO

unidadeOrganica = df_data["UO"].drop_duplicates()

for uo in unidadeOrganica.index:
    unOrg = df_data.loc[uo]["UO"]
    totalPortugueses = df_data.loc[df_data["UO"]==unOrg]["Portugues"].sum()
    totalEstrangeiros = df_data.loc[df_data["UO"]==unOrg]["Estrangeiro"].sum()
    
    print(unOrg)
    print("  -Portugueses: " + str(totalPortugueses))
    print("  -Estrangeiros: " + str(totalEstrangeiros))
```

```python
#Nº total de diplomados por publico/privado

publicoPrivado = df_data["Subsistema_ensino"].drop_duplicates()

for pubPriv in publicoPrivado.index:
    newPubPriv = df_data.loc[pubPriv]["Subsistema_ensino"]
    totalPortugueses = df_data.loc[df_data["Subsistema_ensino"]==newPubPriv]["Portugues"].sum()
    totalEstrangeiros = df_data.loc[df_data["Subsistema_ensino"]==newPubPriv]["Estrangeiro"].sum()
    
    print(newPubPriv)
    print("  -Portugueses: " + str(totalPortugueses))
    print("  -Estrangeiros: " + str(totalEstrangeiros))
```

```python
#Nº total de diplomados publico por distrito e concelho

distritos = df_data["Distrito"].drop_duplicates()
concelhos = df_data["Conselho"].drop_duplicates()

print("\n" "Distrito" "\n")

for dist in distritos.index:
    distrito = df_data.loc[dist]["Distrito"]
    totalPublico = df_data.loc[(df_data["Distrito"]==distrito) & (df_data["Subsistema_ensino"]=="Público") ]["Total"].sum()

    print(distrito)
    print("  -Público: " + str(totalPublico))


print("\n" "Concelho" "\n")
for conc in concelhos.index:
    concelho = df_data.loc[conc]["Conselho"]
    totalPublico = df_data.loc[(df_data["Conselho"]==concelho) & (df_data["Subsistema_ensino"]=="Público") ]["Total"].sum()

    print(concelho)
    print("  -Público: " + str(totalPublico))
```

```python
#Nº total de diplomados privado por distrito e concelho

distritos = df_data["Distrito"].drop_duplicates()
concelhos = df_data["Conselho"].drop_duplicates()

for dist in distritos.index:
    distrito = df_data.loc[dist]["Distrito"]
    totalPrivado = df_data.loc[(df_data["Distrito"]==distrito) & (df_data["Subsistema_ensino"]=="Privado") ]["Total"].sum()

    print(distrito)
    print("  -Privado: " + str(totalPrivado))


print("----------------")
for conc in concelhos.index:
    concelho = df_data.loc[conc]["Conselho"]
    totalPrivado = df_data.loc[(df_data["Conselho"]==concelho) & (df_data["Subsistema_ensino"]=="Privado") ]["Total"].sum()

    print(concelho)
    print("  -Privado: " + str(totalPrivado))
```

```python
#Nº total de diplomados privado por curso e concelho

cursos = df_data["Curso"].drop_duplicates()
concelhos = df_data["Conselho"].drop_duplicates()

for conc in concelhos.index:
    concelho = df_data.loc[conc]["Conselho"]
    print()
    print()
    print(concelho)
    for cursoIndex in cursos.index:
        curso = df_data.loc[cursoIndex]["Curso"]
        diplomados = df_data.loc[(df_data["Conselho"]==concelho) & (df_data["Curso"]==curso) ]["Total"].sum()
        if(diplomados>0):
            print("         -" + str(curso) + "  - " + str(diplomados))
```
